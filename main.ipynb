```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('air_quality.csv',low_memory=False)
```

```python
data.info()
```

```python
data = data.drop(columns=['unit'])
```

```python
data["so2_avg"] = data["so2_avg"].replace('-', 0.0)
```

```python
# Calculate the mean of the non-NaN values in the 'so2_avg' column
mean_so2_avg = data['so2_avg'].astype(float).mean()

# Replace NaN values with the calculated mean
data['so2_avg'] = data['so2_avg'].replace(np.nan, mean_so2_avg)
```
